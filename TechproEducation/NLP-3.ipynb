## Word2Vec

```python
import numpy as np

from gensim.models import Word2Vec
```

We have the dataset which is include 400k news row and already cleaned.

```python
text = open('newspaper.txt', 'r', encoding = 'utf8')       # we should to use encoding = 'utf8' when reading smth.
text = text.read()
list_sent = text.split('\n')

corpus = []

for sent in list_sent:
    corpus.append(sent.split())   # We took all words from all sentences.                      
```

```python
print(corpus[:10])
```

```python
model = Word2Vec(corpus, vector_size = 100, window = 5, min_count = 5, sg = 1)
```

vector_size -->__ Word Embedding Size (Number of Attributes).

__window (window_size) -->__ How many words will it look at to the left and right of the corresponding token?

__min_count=5 -->__ Do not include 5 or less spent tokens in training in Corpus.

__sg (Skip-gram) -->__ sg = 1 --> skip-gram    ,     sg = 0 --> CBOW

CBOW gives better results in large corpuscles and Skip-Gram in small corpuscles. We preferred Skip-Gram because our dataset is small this time.

We built our model above. We give the word2Vec model of which token we want to see. Word Embedding with size 100 has occurred

```python
model.wv['turkey']
```

The __most_similar__ function is used to understand how well a Word Embedding model is trained. It is observed whether the meaning of aggregation in a new feature has changed by removing a feature from two different tokens.

```python
model.wv.most_similar(positive = ['king', 'queen'], negative = ['woman'], topn = 1) # Not working good.
```

Rusya' dan Moskova cikarildiginda ulke kalir, ulkeye Ankara ilave edildiginde 'Turkiye' bulunur. Model bunu yakalayabilmis. Gazete haberlerinde ulke isimleri cok fazla gectigi icin modelin bunu yakalayabilmesi beklenen bir sonuc :

## Glove

https://nlp.stanford.edu/projects/glove/         ----> You can download glove.6B

We will use pre-trained Glove model. To do this, we should 

```python
from gensim.scripts.glove2word2vec import glove2word2vec
from gensim.models import KeyedVectors     
```

We cannot use it directly in the glove model. That's why we will convert the Glove models to Word2Vec format and use it by it.

__KeyedVectors -->__ Glove model is the function we use to call it back from the local after converting it to Word2Vec format and saving it in our local.

The corpus we will use is trained with 6 billion tokens on wikipedia in 2014 (6B) and each token is 100 vector size (100d). We installed our trained glove model over the internet with the name __glove.6B.100d.txt__

```python
glove_model = 'glove.6B.100d.txt'
```

__glove2word2vec -->__ Converts from glove to word2vec and saves to local.

Word Embeddings of each token are fixed as the model has been trained before.

```python
word2vec = 'glove.6B.100d.glove'
glove2word2vec(glove_model, word2vec)
```

We called the model in Word2Vec format that we saved locally with the __KeyedVectors__ function:

```python
model_glove = KeyedVectors.load_word2vec_format(word2vec)
```

Since the model has been trained before, the results are always the same:

```python
model_glove['turkey']
```

The predictions in this model are better than the previous model, as it trains with more words:

```python
model_glove.most_similar('turkey')
```

It is not established according to the lexical meaning, but according to the frequency of using the words together.

```python
model_glove.most_similar('teacher')
```
